## Problem Set-Up

Given a uniformly charged non-conducting sphere, we can find its electric field. The sphere has a total charge, Q, distributed uniformly throughout its volume and radius, R. This scenario allows us to apply Gauss' Law to find the electric field both inside and outside the sphere. By symmetry, the electric field is radial and the magnitude depends only on the distance from the center, r. The charge density $\rho$ is constant throughout the sphere and can be related to the total charge by the volume of the sphere. The goal is to analytically derive the electric field E at any point in space, both inside $(r < R)$ and outside $(r > R)$ the sphere.

In this problem, we are assuming that the spherically symmetric and there is no angular dependence. 

We are also assuming that the system is in electrostatic equilibrium and there is no time-varying elements of the system. Thus, the system can be completely described by Gauss' and Coulomb's Law.

### Variables:

Total charge: $Q$

Radius: $R$

Distance: $r$

Constant ($ 8.854*10^{−12} \frac{C^2}{N m^2}$) : $\epsilon_0$

Charge density: $\rho= \frac{Q}{\frac{4}{3}\pi R^3}$

### Analytical

Using Gauss’s Law we can analytically calculate the electric field both inside and outside the sphere. For $r > R$, the field behaves as if all the charge were concentrated at the center of the sphere, just like a point charge. For $r < R$, the electric field increases linearly with $r$ due to the enclosed charge at that radius.  For a spherically symmetric charge distribution, the volume element is given by $dV=4 \pi r^2 dr$.

Guass's Law: 

$\Phi_E = \oint_{S} E \cdot dA =\frac{Q_{enc}}{\epsilon_0}$

For spherical symmetry, the electric field is constant over the spherical surface, and the area element is $ dA= 4 \pi r^2 $. So, the total flux is:

$\Phi_E=E(4 \pi r^2)$

and

$E(4 \pi r^2)=\frac{Q_{enc}}{\epsilon_0}$.

Calculating the electric field outside the sphere, for $r > R$, 

$\Phi_E = E(4 \pi r^2) $.

Since the entire charge is enclosed, 

$E(4 \pi r^2)= \frac{Q}{\epsilon_0}$

$E=\frac{Q}{\epsilon_0 4 \pi r^2}$

Calculating the electric field inside the sphere for $r < R$, the charge enclosed by a Guassian surface is:

$Q_{enc}(r)= \int_{0}^{r} \rho r' 4 \pi (r')^2 dr' = \frac{Qr^3}{R^3}$

Then using Gauss' Law,

$E (4 \pi r^2)=\frac{Qr^3}{\epsilon_0 R^3}$

$E=\frac{Qr}{ 4 \pi \epsilon_0 R^3}$

Inside the sphere, the electric field increases linearly with the distance from the center.

Outside the Sphere $r→R$:

$E_{outside}(R)= \frac{Q}{4 \pi \epsilon_0}$

Inside the Sphere $r→0$:

$E_{inside}(0)= 0$

So, the electric field is continuous and transitions smoothly from zero at the center to the points outside the sphere.

### Numerical Solution

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#constants
Q = 1e-6  #total charge
R = 0.5  #radius of sphere
epsilon_0 = 8.854e-12

x = np.linspace(-1, 1, 10)
y = np.linspace(-1, 1, 10)
z = np.linspace(-1, 1, 10)
X, Y, Z = np.meshgrid(x, y, z)

#calculate the distance from center
R_sqrt = np.sqrt(X**2 + Y**2 + Z**2)

#calculate electric field components
E = np.zeros_like(R_sqrt)

#electric field outside the sphere
E_outside = np.where(R_sqrt > R, Q / (4 * np.pi * epsilon_0 * R_sqrt**2), 0)

#electric field inside the sphere
E_inside = np.where(R_sqrt <= R, (Q * R_sqrt) / (4 * np.pi * epsilon_0 * R**3), 0)

#total electric field
E = E_outside + E_inside

#normalize the electric fields
Ex = E * (X / R_sqrt)
Ey = E * (Y / R_sqrt)
Ez = E * (Z / R_sqrt)

#plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, Ex, Ey, Ez, length=0.1, normalize=True)
ax.set_title('Electric Field of a Uniformly Charged Sphere')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
plt.show()
```
